```python
from bs4 import BeautifulSoup
import html5lib
```

```python
import requests
URL = "https://www.hanumanttechnology.com/"
r = requests.get(URL)
print(r.content)
```

```python
soup = BeautifulSoup(r.content, 'html5lib')
```

```python
nested_images = soup.find_all('li')

for li in nested_images:
    amp_img = li.find('amp-img')
    if amp_img:
        img = amp_img.find('img')
        if img:
            print(img['src'])
```

```python
tags = soup.find_all('amp-img')
tags
```

```python
tags[0]["src"]
```

```python
links = []
for tag in tags:
  links.append(tag["src"])
```
